```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## y = 2X + 2

```python
xs = np.arange(1, 20, 1.0)
ys = np.array([(2*i+2) for i in xs], dtype=float)
```

```python
xs
```

```python
ys
```

### Build simple model

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(xs, ys, epochs=5050, verbose=1)
```

`Jika dilihat loss masih sangat tinggi`

```python
model.predict([1000.])
```

`Predicsi pun tidak akurat`

### Build simple neural network

`Kalau model sebelumnya hanya single perceptron karena tidak ada hidden layer`

```python
model2 = tf.keras.Sequential([
            tf.keras.layers.Dense(units=1, input_shape=[1]),
            tf.keras.layers.Dense(units=8),
            tf.keras.layers.Dense(units=1)
])
```

```python
model2.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model2.fit(xs, ys, epochs=6000)
```

```python
model2.predict([0.])
```

```python
model2.predict([1000.])
```

`Prediksi lebih akuran`
